# Data loader

```python
import gzip
import pandas as pd
```

```python
df_tweets = pd.read_csv('tweets.csv.gz', compression='gzip', header=0, sep=';', quotechar='"')
df_topics = pd.read_csv('topics.csv.gz', compression='gzip', header=0, sep=';', quotechar='"')
df_polarities = pd.read_csv('polarities.csv.gz', compression='gzip', header=0, sep=';', quotechar='"')
```

```python
df_topics[['tweetid', 'topic']].groupby('topic').count()
```

```python
res = df_tweets.join(df_topics.set_index('tweetid'), on='tweetid').drop_duplicates(subset='tweetid', keep="first")

res[['tweetid', 'topic']].groupby('topic').count()
```
